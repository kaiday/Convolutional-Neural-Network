```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
source_data = 'Pistachio_28_Features_Dataset.xlsx'

xl = pd.ExcelFile(source_data)

df_source = xl.parse("Pistachio_28_Features_Dataset")
df = df_source.drop('Class', axis=1)
df=df.T
```

```python
#Standardlize the data
mean = df.mean(numeric_only=True)
mean
```

```python
Standardlize_data = df-mean
Standardlize_data
```

```python
N=len(df)

S= ( (np.matmul( np.array(Standardlize_data) ,Standardlize_data.T)) ) / N

S
```

```python
eigenvalues, eigenvectors = np.linalg.eig(S)
idx = np.argsort(eigenvalues)[::-1]

eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:,idx]

B=eigenvectors[:,:2]
B.shape
```

```python
z = np.matmul(B.T,Standardlize_data).values

z=z.T
```

```python
fig = plt.figure()
ax = fig.subplots()

ax.scatter(z[df_source['Class'] == 'Kirmizi_Pistachio',0 ], z[df_source['Class'] == 'Kirmizi_Pistachio', 1], c='r', marker='o', label='Class 1')
ax.scatter(z[df_source['Class'] == 'Siirt_Pistachio',0], z[df_source['Class'] == 'Siirt_Pistachio', 1], c='b', marker='^', label='Class 2')
plt.show()
```
