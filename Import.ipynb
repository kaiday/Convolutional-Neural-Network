```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# load the dataset
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()

# convert to pandas DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)

# add target to DataFrame
df['target'] = data.target

df
```

```python
mean_of_each_collumn = df.mean()

pd.DataFrame(mean_of_each_collumn)
```

```python
std_of_each_collumn = df.std()

pd.DataFrame(std_of_each_collumn).T
```

```python
# standardize data
standardize_data = (df - mean_of_each_collumn) / std_of_each_collumn
standardize_data
```

```python
covarance = np.cov(standardize_data.T, bias = 1)
```

```python
eigenvalue, eigenvectors = np.linalg.eig(covarance)
```

```python
n_components = 2

pca_manual = np.matmul(np.array(standardize_data),eigenvectors)

pca_manual  = pca_manual[:,:n_components]

pca_manual
```

```python
fig = plt.figure()
ax = fig.subplots()
ax.scatter(pca_manual[:,0], pca_manual[:,1])

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_manual[:,0],pca_manual[:,1],c=data['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```
