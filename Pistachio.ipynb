```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
source_data = 'Pistachio_28_Features_Dataset.xlsx'

xl = pd.ExcelFile(source_data)

df = xl.parse("Pistachio_28_Features_Dataset")

# Use the replace function to replace values
df['Class'] = df['Class'].replace({'Kirmizi_Pistachio': 0, 'Siirt_Pistachio': 1})
df=(df.T).iloc[:, :100]
df
```

```python
#Standardlize the data
mean = df.mean(numeric_only=True)
mean
```

```python
Standardlize_data = df-mean
Standardlize_data
```

```python
N=len(df)

S= ( (np.matmul( np.array(Standardlize_data) ,Standardlize_data.T)) ) / N

S
```

```python
eigenvalues, eigenvectors = np.linalg.eig(S)
idx = np.argsort(eigenvalues)[::-1]

eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:,idx]

B=eigenvectors[:,:2]
B
```

```python
z = np.matmul(B.T,Standardlize_data)

z.T
```

```python
fig = plt.figure()

(z.T).plot()

plt.show()
```

```python
from sklearn.decomposition import PCA

# Create an instance of PCA
pca = PCA(n_components=2)  # Specify the number of components to retain

# Fit the PCA model to the data and transform the data
transformed_data = pca.fit_transform(df.T)
df1 = pd.DataFrame(transformed_data)

plt.scatter(df1[0], df1[1])

plt.axhline(0, color='black')

plt.axvline(0, color='black')
plt.show()

```
